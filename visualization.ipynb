```python
import os
from dotenv import load_dotenv

# load the environment variables from .env
load_dotenv()
```

```python
revenue_rating_by_years = os.environ.get('revenue_rating_by_years')
num_of_bookings_by_month_and_year = os.environ.get('num_of_bookings_by_month_and_year')
percent_of_bookings_by_status = os.environ.get('percent_of_bookings_by_status')
number_of_declined_top_two_months_by_years = os.environ.get('number_of_declined_top_two_months_by_years')
number_of_cancelation_by_days = os.environ.get('number_of_cancelation_by_days')
number_of_cancelations_by_month = os.environ.get('number_of_cancelations_by_month')
percent_of_tot_book_percent_of_tot_revenue = os.environ.get('percent_of_tot_book_percent_of_tot_revenue')
revenue_by_time = os.environ.get('revenue_by_time')
tot_revenue_avg_rating_by_year_by_month = os.environ.get('tot_revenue_avg_rating_by_year_by_month')
```

```python
# Avg Commission and Avg Rating Over the Years

df = pd.read_csv(revenue_rating_by_years)

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Avg Commission on the primary y-axis (ax1)
ax1.plot(df['theYear'], df['AvgCommission'], color='blue', marker='o', label='Avg Commission')
ax1.set_xlabel('Year')
ax1.set_ylabel('Avg Commission', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a secondary y-axis
ax2 = ax1.twinx()
ax2.plot(df['theYear'], df['AvgRating'], color='green', marker='o', label='Avg Rating')
ax2.set_ylabel('Avg Rating', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Adjust the scale of the secondary y-axis for 'Avg Rating'
ax2.set_ylim(min(df['AvgRating']) - 0.1, max(df['AvgRating']) + 0.1)

plt.xticks(df['theYear'].astype(int))

# Adding title
plt.title('Avg Commission and Avg Rating Over the Years')

# Display the plot
plt.show()
```

```python
df = pd.read_csv(num_of_bookings_by_month_and_year)
```

```python
df["year_month"] = df['Year'].astype(str) + '-' + df['MonthOfYear'].astype(str)
df["year_month"] = pd.to_datetime(df["year_month"], format='%Y-%m')

plt.figure(figsize=(10, 6))

plt.plot(df["year_month"], df["count(*)"], marker='o', linestyle='-', color='b')
plt.xlabel('Year-Month')
plt.ylabel('Count')
plt.title('Count Over Year-Month')

plt.xticks(df['year_month'], rotation=90, fontsize=8) 

plt.gca().xaxis.set_major_formatter(DateFormatter('%Y-%m'))

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv(percent_of_tot_book_percent_of_tot_revenue)
```

```python
bar_width = 0.35
index = range(len(df))

plt.bar(index, df['PercentOfTotalBookings'], width=bar_width, label='PercentOfTotalBookings', color='skyblue')
plt.bar([i + bar_width for i in index], df['PercentOfTotalRevenue'], width=bar_width, label='PercentOfTotalRevenue', color='orange')

plt.xlabel('Commission Level')
plt.ylabel('Percentage')
plt.title('Comparison of PercentOfTotalBookings and PercentOfTotalRevenue')
plt.xticks([i + bar_width/2 for i in index], df['LABEL'], rotation=45)
plt.legend()

plt.show()
```
